```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"/content/student_lifestyle_100k.csv")
```

```python
df.describe()
```

```python
df['Gender'] = df['Gender'].astype('category')
df['Department'] = df['Department'].astype('category')
```

```python
df['Stress_Level'] = df['Stress_Level'].astype('category')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

```python
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
```

```python
df['CGPA'].fillna(df['CGPA'].mean(), inplace=True)
```

```python

df['Sleep_Duration'].fillna(df['Sleep_Duration'].mean(), inplace=True)
```

```python
df['Study_Hours'].fillna(df['Study_Hours'].mean(), inplace=True)
```

```python
df['Social_Media_Hours'].fillna(df['Social_Media_Hours'].mean(), inplace=True)
```

```python
df['Physical_Activity'].fillna(df['Physical_Activity'].mean(), inplace=True)
```

```python
df['Stress_Level'].fillna(df['Stress_Level'].mode()[0], inplace=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df['Depression'] = df['Depression'].fillna(False)
```

```python
sns.boxplot(y=df['Social_Media_Hours'])
plt.show()
```

```python
df['Total_Screen_Time'] = df['Social_Media_Hours'] + df['Study_Hours']
df['Sleep_Category'] = pd.cut(
    df['Sleep_Duration'],
    bins=[0,5,7,24],
    labels=['Low Sleep','Normal Sleep','High Sleep']
)
```

```python
df['Depression'] = df['Depression'].astype(int)
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```

```python
sns.scatterplot(x='Study_Hours', y='CGPA', data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['CGPA'], kde=True, bins=30)
plt.title('Distribution of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Count')
plt.show()
```

Most students have CGPA clustered around the average range, with fewer students at extreme low or high CGPA values.

```python
sns.kdeplot(df['Sleep_Duration'], fill=True)
plt.title('Sleep Duration Density')
plt.show()
```

The majority of students sleep within a normal duration range, while very short or very long sleep durations are less common.

```python
sns.boxplot(x='Gender', y='CGPA', data=df)
plt.title('CGPA by Gender')
plt.show()
```

CGPA distribution is similar across genders, with minor differences and some outliers in both groups.

```python
sns.violinplot(x='Gender', y='Study_Hours', data=df)
plt.title('Study Hours Distribution by Gender')
plt.show()
```

Both genders show a similar study-hour pattern, though one group may have slightly higher study-hour concentration.

```python
sns.barplot(x='Stress_Level', y='CGPA', data=df)
plt.title('Average CGPA by Stress Level')
plt.show()
```

Higher stress levels are generally associated with slightly lower average CGPA.

```python
plt.figure(figsize=(12, 10))
sns.scatterplot(x='Study_Hours', y='CGPA', data=df)
plt.title('Study Hours vs CGPA')
plt.show()
```

Students who study more hours tend to achieve higher CGPA, showing a positive relationship.

The upward trend confirms that increased study hours positively impact academic performance.

```python
sns.pairplot(df[['CGPA','Sleep_Duration','Study_Hours','Social_Media_Hours']])
plt.show()
```

```python
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

CGPA shows positive correlation with study and sleep duration, and negative correlation with excessive social media usage.

```python
sns.countplot(x='Depression', data=df)
plt.title('Depression Distribution')
plt.show()
```

Most students fall into the non-depressed category, though a significant portion shows signs of depression.

```python
pd.crosstab(df['Stress_Level'], df['Depression']).plot(kind='bar', stacked=True)
plt.title('Stress Level vs Depression')
plt.show()
```

Higher stress levels are associated with a higher proportion of depressed students.
